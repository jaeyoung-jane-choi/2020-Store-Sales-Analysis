```python

#loading dataset 

import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_palette("colorblind")
import os
import json
import urllib.request
import pickle
import datetime
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split



%matplotlib inline
fig = plt.figure(figsize=(20,5))

print(os.getcwd())


def load_data(data_dir):
    
    data = pd.read_csv(data_dir)

    #make amount into log value  :: when predicting use np.expm1(amount) to make it back to original amount size
    data['amount'] = np.log(data["amount"]) #log0 error 방지 
    data['three_month'] = np.log(data["three_month"])
    data['four_month'] = np.log(data["four_month"])
    data['five_month'] = np.log(data["five_month"])
    data['six_month'] = np.log(data["six_month"])
    data['seven_month'] = np.log(data["seven_month"])
    data['eight_month'] = np.log(data["eight_month"])
    data['nine_month'] = np.log(data["nine_month"])
    data['ten_month'] = np.log(data["ten_month"])
    data['eleven_month'] = np.log(data["eleven_month"])
    data['twelve_month'] = np.log(data["twelve_month"])
    
    
    #drop store 795
    l = data.loc[(data.store_id == 795)].index
    data.drop(index=l, inplace=True)
    
    
    #one-hot-encoding ##
#     data = pd.get_dummies(data, columns = ['store_id']) 
    #'year','month' : not ordered store_id may need one hot encoding 
    #but year, month has order 
    
    #make values to category 
    data['store_id'] = data['store_id'].astype('category')
    data['month'] = data['month'].astype('category')
    
    #delete category and use train test split to split dataset 
    test = data[data['category']== 'test']
    data =data[data['category']!= 'test']
    
    del data['category']
    del test['category']
#     #split train/val/test
#     train = data[data['category'] == 'train']
#     val = data[data['category']=='val']
    
    X = data[['store_id', 'year', 'month', 'three_month', 
              'four_month','five_month','six_month', 
              'seven_month', 'eight_month', 'nine_month', 
              'ten_month','eleven_month','twelve_month']]
    
    y =data['amount']
    
    X_train, X_val, y_train, y_val = train_test_split(X, y , 

                                                    test_size=0.2, 

                                                    shuffle=True,

                                                    stratify=data['store_id'], 

                                                    random_state=1004)


    
    
    return  X_train, X_val, y_train, y_val,test 
    
    




######NEWNEWNEWNENW########## 

# X_train, X_val, y_train, y_val,test  = load_data('/Users/janechoi/Desktop/연구학점/credit card/data/train_na_full_month.csv')



# X_train, X_val, y_train, y_val,test  = load_data('/Users/janechoi/Desktop/연구학점/credit card/data/train_mean_full_month.csv')

# X_train, X_val, y_train, y_val,test  = load_data('/Users/janechoi/Desktop/연구학점/credit card/data/train_median_full_month.csv')




# X_train, X_val, y_train, y_val,test  = load_data('/Users/janechoi/Desktop/연구학점/credit card/data/train_linear_full_month.csv')

# X_train, X_val, y_train, y_val,test  = load_data('/Users/janechoi/Desktop/연구학점/credit card/data/train_spline2_full_month.csv')

X_train, X_val, y_train, y_val,test  = load_data('/Users/janechoi/Desktop/연구학점/credit card/data/train_time_full_month.csv')



# X_train['store_id'].value_counts().index
X_train.shape
# X_train.isnull().sum()
X_train
```

```python
from catboost import Pool, CatBoostRegressor
from bayes_opt import BayesianOptimization
from sklearn.model_selection import GroupKFold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score

#columns to be used 
avoid_cols = ['amount', 'category','year']
cols = [col for col in X_train.columns if col not in avoid_cols] #store_id, 3,6,12month, month 
cat = ['store_id', 'month']

#how to use CV 
skf = list(StratifiedKFold(n_splits=5,random_state=1008).split(X_train, X_train['store_id'])) 



#USING BAYESIAN OPTIMIZATION !!!!! 

def cat_eval(iterations,learning_rate,depth,l2_leaf_reg,bagging_temperature):
    
    #Make lgbm dataset 
    
#     train_data = lgb.Dataset(data= X_train[cols].values, label=y_train)
#     valid_data = lgb.Dataset(X_val[cols].values, label=y_val)
    train = X_train[cols]
    val = X_val[cols]
    
    #select params to use 
    params = {'random_seed': 314 , 
            'od_type':'Iter', #The type of the overfitting detector to use
            'od_wait':100, #number of iterations to continue the training after the iteration with the optimal metric value.
            'use_best_model' : True,
             'cat_features': cat}

    
    #optimize parametes :: 
    
    params["iterations"] = int(round(iterations))
    params['learning_rate'] = learning_rate
    params['depth'] = int(round(depth))
    params['l2_leaf_reg'] = max(l2_leaf_reg, 0)
    params['bagging_temperature'] = int(round(bagging_temperature))

    
    #use cv to get result 
    
    best=[]
    res = CatBoostRegressor(**params)
    #use skf for 5 fold seperatly 
    for fold_, (trn_, val_) in enumerate(skf):
        
        print("Fold:",fold_)
        
        
        train_x, train_y = train.iloc[trn_], y_train.iloc[trn_]
        other_x, other_y = train.iloc[val_], y_train.iloc[val_]
    
        res.fit(train_x, train_y ,eval_set = [(val, y_val)],verbose=False)
    
        pred= res.predict(other_x)
        
        #calculate mae on other train fold 
        mae_error = mean_absolute_error(np.exp(other_y), np.exp(pred))
    
        best.append(mae_error)
        
#     print(best)
    mean = np.mean(np.array(best))
#     print(mean)
    return  -(mean) #maximize :: minimize minus mae 



catBO = BayesianOptimization(cat_eval, {'iterations': (1000,2000) ,
                                        'learning_rate' :(0.01, 0.1), 
                                        'depth': (5,16),
                                        'l2_leaf_reg':  (0,100),
                                        'bagging_temperature' : (0,100)
                                        },
                                        random_state= 0 )


#randomly use 10 initial points, and use bayesian for 1000 more rounds 
catBO.maximize(init_points=5, n_iter=50)  
catBO.max['params'] 


```

```python
from sklearn.metrics import mean_absolute_error

# {'bagging_temperature': 0.3637754563158624,
#  'depth': 10.225769823638064,
#  'iterations': 1627.7507217886825,
#  'l2_leaf_reg': 17.801749568451374,
#  'learning_rate': 0.05091659983967498}

def cat_cv(c=None,v=None):
    
    #columns to be used 
    avoid_cols = ['amount', 'category','year',c,v]
    cols = [col for col in X_train.columns if col not in avoid_cols] #store_id, 3,6,12month, month 
    cat = ['store_id', 'month']
    
    train = X_train[cols]
    val = X_val[cols]
    
    
    #basic parameters 
    basicparams = {'random_seed': 314 , 
            'od_type':'Iter', #The type of the overfitting detector to use
            'od_wait':100, #number of iterations to continue the training after the iteration with the optimal metric value.
            'use_best_model' : True,
             'cat_features': cat}

    
    #optimized parametes update 
    params = catBO.max['params']
    
    k = params['iterations']
    params['iterations'] = int(round(k))
    
    k = params['learning_rate']
    params['learning_rate'] = k
    
    k = params['depth']
    params['depth'] = int(round(k))
    
    k = params['bagging_temperature']
    params['bagging_temperature'] = int(round(k))
    
    k = params['l2_leaf_reg']
    params['l2_leaf_reg'] = k
    
    
    params.update(basicparams)
    
    #fit on catbosst regressor 
    train_x =X_train[cols]
    val = X_val[cols]
    
    res = CatBoostRegressor(**params)
    res.fit(train_x, y_train ,eval_set = [(val, y_val)],verbose=False)
    
    #predictions 
    predictions = res.predict(X_val[cols])
    error= mean_absolute_error(np.exp(y_val), np.exp(predictions))
    
    print('MAE for validation set is...' + str(error))

    
    return cols, res 
    

cols,m = cat_cv()



#MAE for NA validation set is...298405.24679300195
# feature excluded...three_month
# MAE for validation set is...300377.0331641792
# feature excluded...four_month
# MAE for validation set is...299817.3676271339
# feature excluded...five_month
# MAE for validation set is...299635.0545106586
# feature excluded...six_month
# MAE for validation set is...300497.97129494103
# feature excluded...seven_month
# MAE for validation set is...307946.29184894596
# feature excluded...eight_month
# MAE for validation set is...301833.8398858013
# feature excluded...nine_month
# MAE for validation set is...299332.9296058349

# feature excluded...ten_month :::::::BEST 
# MAE for validation set is...296882.5269922852

# feature excluded...eleven_month
# MAE for validation set is...297539.1149293383
# feature excluded...twelve_month
# MAE for validation set is...298979.531469938


###MEAN 
# MAE for validation set is...307152.91298629215

# feature excluded...three_month
# MAE for validation set is...307863.6040581839
# feature excluded...four_month
# MAE for validation set is...296820.90728529356
# feature excluded...five_month
# MAE for validation set is...298420.0564820828
# feature excluded...six_month
# MAE for validation set is...303824.91455985914
# feature excluded...seven_month
# MAE for validation set is...291090.20268030994
# feature excluded...eight_month
# MAE for validation set is...290323.3667811219


# feature excluded...nine_month :::::BEST 
# MAE for validation set is...289726.04744217475


# feature excluded...ten_month
# MAE for validation set is...298412.0150377637
# feature excluded...eleven_month
# MAE for validation set is...302142.6206009675
# feature excluded...twelve_month
# MAE for validation set is...292688.77751730126



###MEDIAN
# MAE for validation set is...298237.7863820528
# feature excluded...three_month
# MAE for validation set is...297445.218314663
# feature excluded...four_month
# MAE for validation set is...302978.9040500115


# feature excluded...five_month ::BSET
# MAE for validation set is...288216.2977247809


# feature excluded...six_month
# MAE for validation set is...295548.8639145098
# feature excluded...seven_month
# MAE for validation set is...300598.5450871778
# feature excluded...eight_month
# MAE for validation set is...291061.7611532172
# feature excluded...nine_month
# MAE for validation set is...297568.4322875917
# feature excluded...ten_month
# MAE for validation set is...295508.72293859173
# feature excluded...eleven_month
# MAE for validation set is...294880.9879797481
# feature excluded...twelve_month
# MAE for validation set is...295470.25477660954

###LINEAR
# MAE for validation set is...288411.4708037652
# feature excluded...three_month
# MAE for validation set is...318048.26539080293
# feature excluded...four_month
# MAE for validation set is...292809.93098431197
# feature excluded...five_month
# MAE for validation set is...290002.74307629943
# feature excluded...six_month
# MAE for validation set is...295751.56261544535

# feature excluded...seven_month :::::::BEST 
# MAE for validation set is...286766.30711342575


# feature excluded...eight_month
# MAE for validation set is...294899.6782626636
# feature excluded...nine_month
# MAE for validation set is...290354.4358211517
# feature excluded...ten_month
# MAE for validation set is...292053.46331128007
# feature excluded...eleven_month
# MAE for validation set is...288690.2100633133
# feature excluded...twelve_month
# MAE for validation set is...290557.3785314451


###SPLINE

###TIME 
# MAE for validation set is...291848.42805239826
# feature excluded...three_month
# MAE for validation set is...310634.13878635596
# feature excluded...four_month
# MAE for validation set is...295487.7997162063
# feature excluded...five_month
# MAE for validation set is...288577.4341099604
# feature excluded...six_month
# MAE for validation set is...296123.37416375906
# feature excluded...seven_month
# MAE for validation set is...288975.6045408248

# feature excluded...eight_month ::BEST 
# MAE for validation set is...287251.3764865681

# feature excluded...nine_month
# MAE for validation set is...290258.57691374974
# feature excluded...ten_month
# MAE for validation set is...290689.01595232467
# feature excluded...eleven_month
# MAE for validation set is...291918.36491846957
# feature excluded...twelve_month
# MAE for validation set is...299636.3347294214







```
